```python
import pandas as pd
import glob
import os

ut_enrollment_2020 = pd.read_csv("data/ut_2020-2021_enrollment_school.csv")
df = ut_enrollment_2020
df.head()
```

```python
# remove schools that are not high school
df.drop(columns=['K', 'Grade_1', 'Grade_2', 'Grade_3', 'Grade_4', 'Grade_5', 'Grade_6', 'Grade_7'])
```

```python
# drop schools that do not have any students in grades 8-12 
df = df[df.Grade_8.notnull() and df.Grade_9.notnull() and df.Grade_10.notnull() and df.Grade_11.notnull() and df.Grade_12.notnull()]
```

```python
# print school names and subgroups
ut_school_name = df['School Name'].unique()
print("Unique schools", ut_school_name)
```

```python
# getting the number of unique values for school names 
ut_schools = len(pd.unique(df['School Name']))
print('Utah Schools', ut_schools)
```
