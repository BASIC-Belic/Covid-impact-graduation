Answer: There is a supplementary file called codebook_districts_ccd_finance.xls that provides a way to convert the "fips" column to a Statename. Once this is added we can groupby state and year and run any state level queries as needed such as sums, aggregates etc. 

Provided below are the steps using the data from https://educationdata.urban.org/documentation/school-districts.html#ccd_finance

````python
import pandas as pd

finance_df = pd.read_csv('finance_data/districts_ccd_finance.csv.zip')
print(finance_df.columns)
finance_df
````

````python
# Adding State column by reading codebook
value_mappings = pd.read_excel(open('finance_data/codebook_districts_ccd_finance.xls', 'rb'),
              sheet_name='values')

fips_codes = value_mappings[value_mappings["format"] == "fips"].set_index('code').code_label.to_dict()
finance_df["state_name"] = finance_df["fips"].map(fips_codes)
print(finance_df.columns)
finance_df
````

````python
# Group by state and year and aggregate a few columns.
# 

finance_df_grouped = finance_df.groupby(['state_name', "year"]).agg(
    {'assets_bond_fund': ['sum'],
    'rev_total': ['sum'],
    'rev_fed_total': ['sum'], 
    'debt_longterm_retired_FY': ['sum'],
    'debt_longterm_outstand_end_FY': ['sum'] }).reset_index()


finance_df_grouped
# finance_df_grouped.assets_bond_fund.sum 
````
